### Question 8

Create a stacked bar chart showing the top 7 UFO shapes for the ten states with the highest number of sightings. Start by finding the ten states with the highest number of sightings. Given those 10 states, find the top 7 most frequent UFO shapes. Then for each state plot a stacked bar where each color of the stack corresponds to the number of sightings in that state for a particular UFO shape. Label the x-axis with the State abbreviation. 

There is an [example](https://matplotlib.org/3.2.2/gallery/lines_bars_and_markers/bar_stacked.html) of a stacked bar chart available in the documentation. 

Use the color brewer website from the class slides to pick 7 colors to override the matplotlib defaults for the bars.

```python
#find the top 10 states with the highest number of sightings 
sightings_by_states = df_sightings[df_sightings['Country'] == 'USA ']['State'].value_counts()
top_10_states = sightings_by_states[0:10]
df_top_states = df_sightings[df_sightings['State'].isin(top_10_states.index)]
df_top_states
```

```python
#create dataframe of sightings of the top 7 most frequent UFO shape among top 10 states
top_shapes = df_top_states['Shape'].value_counts()[0:7]
df_top_shapes = df_top_states[df_top_states['Shape'].isin(top_shapes.index)]
df_top_shapes
```

```python
#create dataframe with counts of each state-shape sighting 
df_state_shape_count = df_top_shapes.groupby(['State','Shape']).count().reset_index()

#sort dataframe by state as ordered categorical data
states_ordered = top_10_states.index.to_list()
df_state_shape_count['State'] = pd.Categorical(df_state_shape_count['State'], categories = states_ordered)
df_state_shape_count = df_state_shape_count.sort_values(by=['State']).reset_index(drop=True)
df_state_shape_count
```

```python
#plot stacked bar chart
fig, ax = plt.subplots(figsize=(10,6))

#top 7 shapes in order
shape_list = df_state_shape_count.groupby('Shape').sum().sort_values('Location',ascending=False).index.to_list()

#list of colors
colors = ['#feebe2','#fcc5c0','#fa9fb5','#f768a1','#dd3497','#ae017e','#7a0177']

#initialize bottom at zero for y-coordinate for first stack
b = np.zeros(10)

#create loop to plot each stack
for i in range(len(shape_list)):
    shape_stack = df_state_shape_count[df_state_shape_count['Shape'] == shape_list[i]]['Location']
    ax.bar(top_10_states.index, shape_stack, bottom = b, color=colors[i], label= shape_list[i])
    b += np.array(shape_stack)

#add chart labels
ax.set_ylabel('Number of Sightings')
ax.set_xlabel('US State')
ax.set_title('UFO Sizes in the Top 10 Most Common States')
ax.legend()
```
